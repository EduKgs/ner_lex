### 连接数据库

```python
import pymysql
import pandas as pd
import gc
from sqlalchemy import create_engine

read_con = pymysql.connect(host="192.168.126.124",port = 3336, user='root', password='123456', database='ccks_2019',charset='utf8')

save_con = create_engine('mysql+pymysql://root:123456@192.168.126.124:3336/ccks_2019?charset=utf8')
cur = read_con.cursor()
```

### 将pkubase-triples导入数据库

```python
sql = """CREATE TABLE `pkubase` (
	`id` INT(11) NOT NULL AUTO_INCREMENT,
	`entry` VARCHAR(102) NOT NULL,
	`prop` VARCHAR(46) NOT NULL,
	`value` VARCHAR(320) NOT NULL,
	PRIMARY KEY (`id`),
	INDEX `entry` (`entry`),
	INDEX `prop` (`prop`),
	INDEX `value` (`value`))
COLLATE='utf8_general_ci'
ENGINE=MyISAM
AUTO_INCREMENT= 0
"""
# 创建pkubase
cur.execute(sql)
```

```python
import re
from tqdm import tqdm
pattern = re.compile(r'[\t ]+')
count = 0
other = 0
f =  open('/data/sqzhang/HLT_CCKS_2019/data/官方数据/pkubase-triples-2019/pkubase-triples.txt','r', encoding = 'utf-8').readlines()
triples = []
i = 1
for line in tqdm(f):
    if line.strip()[-1] == ".": 
        temp = line.strip()[0:-1].strip()
    else:
        print("error!",line)
        continue
    if(temp.count('\t') != 2):
        temp = re.sub(pattern, '\t', temp)
    count += 1
    if(temp.count('\t') != 2):
        print(line, "因为格式不符被丢弃")
        other += 1
        continue
    triples.append(temp.split('\t'))
    if(i % 10000 == 0):
        df = pd.DataFrame(triples, columns=["entry", "prop", "value"])
        df.to_sql(name = "pkubase", con = save_con, if_exists = "append", chunksize = 5000, index = False)
        triples = []
    i += 1
df = pd.DataFrame(triples, columns=["entry", "prop", "value"])
df.to_sql(name = "pkubase", con = save_con, if_exists = "append", chunksize = 5000, index = False)
print('共有数据：', count)
print('被丢弃数据：', other)
```

### 导入pkubase-types数据

```python
sql = """CREATE TABLE `pkutype` (
	`id` INT(11) NOT NULL AUTO_INCREMENT,
	`entry` VARCHAR(302) NOT NULL,
	`prop` VARCHAR(8) NOT NULL,
	`value` VARCHAR(39) NOT NULL,
	PRIMARY KEY (`id`),
	INDEX `entry` (`entry`),
	INDEX `prop` (`prop`),
	INDEX `value` (`value`))
COLLATE='utf8_general_ci'
ENGINE=MyISAM
AUTO_INCREMENT= 0
;"""
# 创建 pkutype
cur.execute(sql)
```

```python
import re
pattern = re.compile(r'[\t ]+')
count = 0
other = 0
with open('/data/sqzhang/HLT_CCKS_2019/data/官方数据/pkubase-types-2019/pkubase-types.txt','r', encoding = 'utf-8') as f:
    line = True
    while line:
        ans = []
        for i in range(10000):
            line = f.readline()
            if not line:  # 到 EOF，返回空字符串，则终止循环
                break# 逐行读取  
            count += 1
            temp = line.strip()[0:-1].strip()
            if(temp.count('\t') != 2):
                temp = re.sub(pattern, '\t', temp)
            if(temp.count('\t') != 2):
                print(line, "因为格式不符被丢弃")
                other += 1
                continue
            ans.append(temp.split("\t"))
        df  = pd.DataFrame(ans,columns=["entry","prop","value"])
        df.to_sql(name="pkutype",con = save_con ,if_exists="append",chunksize=5000,index=False)
        del df,ans
        gc.collect()
    print('共有数据：', count)
    print('被丢弃数据：', other)
```
